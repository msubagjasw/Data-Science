```python
genre_movie = movies_df[['id', 'title']]
```

```python
genre_movie['genres'] = movies_df['genres'].str.split().tolist()
genre_movie.head()
```

```python
genre_movie['genres'].isnull().sum()
```

```python
genre_movie['genres'].dropna(inplace=True)
```

```python
genre_movie['title'].isnull().sum()
```

```python
list_genres = []
for lst in genre_movie['genres'].tolist():
    for tsl in lst:
        list_genres.append(tsl)
     
```

```python
list_genres
```

```python
#finding list of genres with set function
list_genres=set(list_genres)
list_genres
```

```python
list_genres= list(list_genres)
```

```python
list_genres
```

```python
for gnr in list_genres :
    genre_movie[gnr] = gnr
```

```python
genre_movie.head()
```

```python
def hot_encoding (nama_df, value_encoding, list_encoding ):
    for M in list_encoding :
        for L in range(len(nama_df[value_encoding])) :  
            if nama_df[M][L] in nama_df[value_encoding][L]:
                nama_df[M][L] = 1
            else :
                nama_df[M][L] = 0
```

```python
hot_encoding (genre_movie, 'genres', list_genres )
```

```python
#hot_encoders
for M in list_genres :
    for L in range(len(genre_movie['genres'])) :  
        if genre_movie[M][L] in genre_movie['genres'][L]:
            genre_movie[M][L] = 1
        else :
            genre_movie[M][L] = 0
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
print("Enter your favorite movies :")
input_movie = input()
arr=[]
if input_movie in genre_movie['title'].tolist() :
    for y in range(len(genre_movie['title'])):
        arr.append(cosine_similarity(genre_movie[genre_movie['title']==input_movie].iloc[:, 3:-1], [genre_movie.iloc[y, 3:-1]]))
    genre_movie['cosine'] = arr
    display(genre_movie.sort_values(['cosine'], ascending=False)[['title', 'cosine']].head(20))
    
else :
    print('Enter another movie!')
```
