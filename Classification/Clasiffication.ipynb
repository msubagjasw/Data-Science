```python
class BasicInfo:
    ''' Just insert dataframe and show all basic information, for running execute 
    BasicInfo(your df).all_call()'''
    
    def __init__(self, df):
        self.df = df
    def heads(self):
        display(self.df.head())
    def shape(self):
        print('Shape of Dataframe :')
        print(self.df.shape)
        print()
    def cols(self):
        print('Columns of Dataframe :')
        print(self.df.columns)
        print()
    def typess(self):
        print('Types information of Data Frame :')
        display(self.df.dtypes.to_frame())
        print()
    def describe(self):
        print('Descriptive Statistic of Dataframe :')
        display(self.df.describe(include='all'))
        print()
    def correlation_plot(self):
        print('Correlation Heatmap :')
        display(sns.heatmap(self.df.corr()))
        print()
    def correlation(self):
        print('Correlation :')
        display(self.df.corr())
        print()
    def isna(self):
        print('Null information :')
        display(self.df.isnull().sum().to_frame())
        print()
    def plot_isna(self):
        print('Plotting Null :')
        #display(sns.heatmap(self.df.isnull(), cbar=False))
        display(msno.matrix(self.df))
        print()
    def all_call(self):
        self.heads()
        self.shape()
        self.cols()
        self.typess()
        self.describe()
        self.correlation_plot()
        self.correlation()
        self.isna()
        self.plot_isna()
```

```python
def hyper_dt(X, y, size):
    #Pipeline
    XTRAIN, XTEST, YTRAIN, YTEST = train_test_split(X, y, test_size=size)
    pipe_dt = Pipeline(steps=[('scaler', StandardScaler()), ('DT', DecisionTreeClassifier())])
    # Parameter Tuning
    param_dist_dt = {"DT__criterion": ['gini', 'entropy'], "DT__max_depth": [6, 10, 20, 50]}
    model = GridSearchCV(pipe_dt, param_grid=param_dist_dt, cv = 5, n_jobs=-1)
    model.fit(XTRAIN, YTRAIN)
    training_acc = accuracy_score(YTRAIN, (model.predict(XTRAIN)))
    return (model, XTRAIN, XTEST, YTRAIN, YTEST, training_acc, model.best_score_)
```

```python
def hyper_logreg(X, y, size):
    #Pipeline
    XTRAIN, XTEST, YTRAIN, YTEST = train_test_split(X, y, test_size=size)
    pipe_logreg = Pipeline(steps=[('scaler', StandardScaler()), ('logreg', LogisticRegression())])
    # Parameter Tuning
    param_dist_logreg = {'logreg__C':[1.0, 0.1, 0.01], 
                 'logreg__solver':['newton-cg', 'lbfgs', 'liblinear'],
                  'logreg__max_iter':[100, 1000]
                 }
    
    model = GridSearchCV(pipe_logreg, param_grid=param_dist_logreg, cv = 5, n_jobs=-1)
    model.fit(XTRAIN, YTRAIN)
    training_acc = accuracy_score(YTRAIN, (model.predict(XTRAIN)))
    return (model, XTRAIN, XTEST, YTRAIN, YTEST, training_acc, model.best_score_)
```

```python
def hyper_svm(X, y, size):
    #Pipeline
    XTRAIN, XTEST, YTRAIN, YTEST = train_test_split(X, y, test_size=size)
    pipe_svm = Pipeline(steps=[('scaler', StandardScaler()), ('svm', svm.SVC())])
    # Parameter Tuning
    param_dist_svm = {'svm__C':[1.0, 0.1, 0.01], 'svm__gamma':[1,0.1,0.01], 'svm__kernel': ['rbf', 'poly', 'sigmoid']}
                 
    
    model = GridSearchCV(pipe_svm, param_grid=param_dist_svm, cv = 5, n_jobs=-1)
    model.fit(XTRAIN, YTRAIN)
    training_acc = accuracy_score(YTRAIN, (model.predict(XTRAIN)))
    return (model, XTRAIN, XTEST, YTRAIN, YTEST, training_acc, model.best_score_)
```

```python
def hyper_knn(X, y, size):
    #Pipeline
    XTRAIN, XTEST, YTRAIN, YTEST = train_test_split(X, y, test_size=size)
    pipe_knn = Pipeline(steps=[('scaler', StandardScaler()), ('knn', KNeighborsClassifier())])
    # Parameter Tuning
    param_dist_knn = {'knn__leaf_size':[30, 50, 100], 'knn__n_neighbors':list(range(1,50)), 'knn__weights': ['uniform', 'distance']}
                 
    
    model = GridSearchCV(pipe_knn, param_grid=param_dist_knn, cv = 5, n_jobs=-1)
    model.fit(XTRAIN, YTRAIN)
    training_acc = accuracy_score(YTRAIN, (model.predict(XTRAIN)))
    return (model, XTRAIN, XTEST, YTRAIN, YTEST, training_acc, model.best_score_)
```

```python
def hyper_rf(X, y, size):
    #Pipeline
    XTRAIN, XTEST, YTRAIN, YTEST = train_test_split(X, y, test_size=size)
    pipe_rf = Pipeline(steps=[('scaler', StandardScaler()), ('rf', RandomForestClassifier())])
    # Parameter Tuning
    param_dist_rf = {"rf__max_depth": [6, 10, 20, 50], "rf__n_estimators": np.array([100,1000]), \
                  'rf__criterion': ['gini', 'entropy']}
    
    model = GridSearchCV(pipe_rf, param_grid=param_dist_rf, cv = 5, n_jobs=-1)
    model.fit(XTRAIN, YTRAIN)
    training_acc = accuracy_score(YTRAIN, (model.predict(XTRAIN)))
    return (model, XTRAIN, XTEST, YTRAIN, YTEST, training_acc, model.best_score_)
```

```python
def hyper_xgb(X, y, size):
    #Pipeline
    XTRAIN, XTEST, YTRAIN, YTEST = train_test_split(X, y, test_size=size)
    pipe = Pipeline(steps=[('scaler', StandardScaler()), ('XGB', XGBClassifier())])
    # Parameter Tuning
    param_dist = {"XGB__max_depth": [6, 10, 20, 50], "XGB__n_estimators": [100, 1000], "XGB__learning_rate": [0.1, 0.01]}
    
    model = GridSearchCV(pipe, param_grid=param_dist, cv = 5, n_jobs=-1)
    model.fit(XTRAIN, YTRAIN)
    training_acc = accuracy_score(YTRAIN, (model.predict(XTRAIN)))
    return (model, XTRAIN, XTEST, YTRAIN, YTEST, training_acc, model.best_score_)
```

```python
#Neural Network
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical 
from keras.wrappers.scikit_learn import KerasClassifier

#Try with neural Network
# baseline model
def create_baseline():
    # create model
    model = Sequential()
    model.add(Dense(64, input_dim = 13, activation='relu'))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(2, activation='softmax'))
    # Compile model
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model
    
```

```python
def hyper_nn(X,y,size):
    XTRAIN, XTEST, YTRAIN, YTEST = train_test_split(X, y, test_size=size)
    model = KerasClassifier(build_fn=create_baseline)
    params = {'nn__epochs':[10, 50, 100], 'nn__batch_size':[10, 32, 64, 100]}
    pipe = Pipeline(steps=[('scaler', StandardScaler()), ('nn', model)])
    grid_search = GridSearchCV(pipe, param_grid=params, cv =5)
    
    grid_search.fit(XTRAIN, to_categorical(YTRAIN))
    training_acc = accuracy_score(YTRAIN, (grid_search.predict(XTRAIN)))
    return (grid_search, XTRAIN, XTEST, YTRAIN, YTEST, training_acc, grid_search.best_score_)
```

```python
nn60 = hyper_nn(X, y, 0.4)
nn70 = hyper_nn(X, y, 0.3)
nn75 = hyper_nn(X, y, 0.25)
nn80 = hyper_nn(X,y, 0.2)
```
